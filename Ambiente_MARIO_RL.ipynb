```python
import gym  
from nes_py.wrappers import JoypadSpace
import gym_super_mario_bros
from gym_super_mario_bros.actions import SIMPLE_MOVEMENT
```

```python

ambiente = gym.make('SuperMarioBros-v1', apply_api_compatibility=True, render_mode='human')
# pegando as movimentações do personagem
ambiente = JoypadSpace(ambiente, SIMPLE_MOVEMENT)
print(ambiente.action_space)
```

```python
observando_estado = ambiente.reset()
for step in range(5000): #carregando 5000 episódios
    acao = ambiente.action_space.sample() #escolhendo uma ação aleatória
    observando_estado, recompensa, terminou, truncated, info = ambiente.step(acao)
    if recompensa < 0:
        print(f"-> Recompensa negativa detectada ({recompensa}). O Mario provavelmente travou. Reiniciando...")
        observando_estado, info = ambiente.reset()
        continue # Pula para o próximo passo, reiniciando o loop
    if terminou or truncated:
        observando_estado = ambiente.reset()
        break
ambiente.close()
```
